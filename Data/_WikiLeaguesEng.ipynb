```python
import urllib.request
from bs4 import BeautifulSoup
import re
from tqdm import tqdm
```

```python
r = urllib.request.urlopen('https://www.wikipedia.org/wiki/Geography_of_association_football')
```

```python
result = r.read()
```

```python
content = result.decode('utf-8')
```

```python
soup = BeautifulSoup(content, 'lxml')
```

```python
skip = ['Heren, Robert', 'Mulinhouse, Louis', 'FIFA 100', 'FIFA National Team Ranking', 'FIFA 51st Congress', 'Garcia Report', 
         'FIFA World Cup (trophy)', 'FIFA History', 'Football']
```

```python
leagues_dict = {}
for tr in soup.find_all('tr'):
    tds = tr.find_all('td')
    #print(tds)
    leagues_list = []
    for index, td in enumerate(tds):
        #print(index)
        #print(td)
        if (index == 0):
            TitleTag = td.find('a')
            if TitleTag:
                country = td.find('a').get('title')
                print(country)
                if country not in skip:
                    good_country = country
        #if index == 3:
            #clubs = td.find('a').get('href')
            #print(clubs)
        if index == 5:
            leagues = td.find_all('a')
            for league in leagues:
                #print(league.getText())
                leagues_list.append(league.getText())
    
        leagues_dict[good_country] = leagues_list
```
