```python
import urllib.request
from bs4 import BeautifulSoup
import re
from tqdm import tqdm 
```

```python
r = urllib.request.urlopen("https://ru.wikipedia.org/wiki/%D0%A4%D1%83%D1%82%D0%B1%D0%BE%D0%BB_%D0%B2_%D0%BC%D0%B8%D1%80%D0%B5")
```

```python
result = r.read()
```

```python
content = result.decode('utf-8')
```

```python
soup = BeautifulSoup(content, 'lxml') 
```

```python
skip = ['Герен, Робер', 'Мулинхаус, Луи', 'ФИФА 100', 'Рейтинг сборных ФИФА', '51-й конгресс ФИФА', 'Доклад Гарсии', 
        'Кубок мира ФИФА (трофей)', 'История ФИФА', 'Футбол']
```

```python
leagues_dict = {}
for tr in soup.find_all('tr'):
    tds = tr.find_all('td')
    leagues_list = []
    for index, td in enumerate(tds):
        if index == 0:
            country = td.find('a').get('title')
            if country not in skip:
                good_country = country
        #if index == 3:
            #clubs = td.find('a').get('href')
            #print(clubs)
        if index == 5:
            leagues = td.find_all('a')
            for league in leagues:
                #print(league.getText())
                leagues_list.append(league.getText())
    
        leagues_dict[good_country] = leagues_list
                
```
