### Маски в tip_name

```python
unique_tips = df.tip_name.unique().tolist()
```

```python
unique_tips
```

```python
#все тип неймы как строки
string_unique = '\n'.join([str(elem) for elem in unique_tips])
```

```python

#удаляем названия команд
def remove_commands(text):
    return re.sub(r'\«.+\»', '', text)
```

```python
def remove_numbers(text):
    return re.sub(r'\(*[0-9]*\)*', '', text)
```

```python
def remove_punctuation(text):
    return re.sub(r'[\:\.\,\-\+\–\|\%\&\#\<\>]', '', text)
```

```python
def remove_eng(text):
    return re.sub(r'[A-Za-z]', '', text)
```

```python
commands = remove_commands(string_unique)
```

```python
numbers =remove_numbers(commands)
```

```python
punctuation = remove_punctuation(numbers)
```

```python
eng = remove_eng(punctuation)
```

```python
print(eng)
```

```python
#датафрейм из новых тип неймов
df1 = pd.DataFrame([x.split(';') for x in eng.split('\n')])

df1.columns=["tip","t", "y"]
df1 = df1.drop(["t", "y"], axis=1)
print(df1[:50])
```

```python
#удаляем дубликаты
df1.drop_duplicates(keep = False, inplace = True) 
```
