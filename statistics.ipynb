```python
df.match_date_time.unique().shape
```

```python
grouped = df.groupby([df.match_date_time, df.team_of_season_1_name_in_russian, df.team_of_season_2_name_in_russian]).count()
grouped
```

Максимальное количество прогнозов на один матч:

```python
grouped.odd.agg('max')
```

Среднее количество прогнозов на один матч:

```python
grouped.odd.agg('mean')
```
